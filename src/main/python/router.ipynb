```python
# read csv file from /Users/paulh/hlrn-cluster/rust-pt-routing/parallel-qsim-berlin/output/v6.4/10pct/routing-sim64_hor600_w4_r192_10pct

import pandas as pd
df = pd.read_csv('/Users/paulh/hlrn-cluster/rust-pt-routing/parallel-qsim-berlin/output/v6.4/10pct/routing-sim192_hor600_w4_r192_10pct/routing-profiling-2025-12-16_18-15-30.csv')
df.head()
```

```python
# print statistics about duration_ns column
print((df['duration_ns']/1e6).describe())

# print 10, 20, 30, ... quantiles of duration_ns column in df
print((df['duration_ns']/1e6).quantile([r/10 for r in range(0,10)]))

#print distribution of duration_ns column with seaborn
import seaborn as sns
sns.displot(df['duration_ns']/1e6)
# add x axis label "Duration (ms)"
import matplotlib.pyplot as plt
plt.xlabel('Duration (ms)')
```

```python
# plot now on x axis and duration_ns on y axis
sns.lineplot(x=df.index, y=df['duration_ns']/1e6)
# add x axis label "Simulation Step"
plt.xlabel('Simulation Step')
# add y axis label "Duration (ms)"
plt.ylabel('Duration (ms)')
plt.yscale(value="log")
# add title "Routing Duration over Simulation Steps"
plt.title('Routing Duration over Simulation Steps')
```

```python
ax = sns.histplot(
    data=df,
    x=df["now"],
    y=df["duration_ns"] / 1e6,
    bins=(1000, 1000),
    cbar=True,
    cbar_kws={"label": "Count"}
)

ax.set_yscale("log")
ax.set_ylabel("Duration (ms)")
```
